```python
from settlement_chat import SettlementChat
```

```python
pm = SettlementChat(log_file="D:\\Workspace\\DevelopmentTeam\\logs\\pm.log",
                    config_file="D:\\Workspace\\DevelopmentTeam\\CONFIG_LIST.json",
                    system_message_file="D:\\Workspace\\DevelopmentTeam\\prompts\\pm.txt",
                    docs_path="D:\\Workspace\\DevelopmentTeam\\rag\\PM")
pm.run("""
    Create the product details and requierments document for
    notification service, which will load data from some db
    and will schedule nofitication base on the loaded records.
    There should be ability for several dbs integration like cvs and sql.
    and there should be ability for scheduling notifications with several diffrent channels like emails and sms.
    the notification calculation logic should be abstract and able to customized by the user.
""")
```

```python
product_description = pm.get_summary()
```

```python
arch = SettlementChat(log_file="D:\\Workspace\\DevelopmentTeam\\logs\\arch.log",
                    config_file="D:\\Workspace\\DevelopmentTeam\\CONFIG_LIST.json",
                    system_message_file="D:\\Workspace\\DevelopmentTeam\\prompts\\architect.txt",
                    docs_path="D:\\Workspace\\DevelopmentTeam\\rag\\Architecture")
arch.run("""
    According to the product details, provide the software architecture for the metioned product.
    consider the project components, project files structure, relevant libraries,
    and any additional necessary information for developent team.
    here are the product details:
"""
+ product_description)
```

```python
product_arch = arch.get_summary()
```

```python
coder = SettlementChat(log_file="D:\\Workspace\\DevelopmentTeam\\logs\\coder.log",
                    config_file="D:\\Workspace\\DevelopmentTeam\\CONFIG_LIST.json",
                    system_message_file="D:\\Workspace\\DevelopmentTeam\\prompts\\coder.txt",
                    docs_path="D:\\Workspace\\DevelopmentTeam\\rag\\Coder")
coder.run("""
    According to the product description,
    Provide full python code implementation for the product,
    Focus on only providing the python files content.
    Here is the product architecture:
"""
+ product_arch)
```
